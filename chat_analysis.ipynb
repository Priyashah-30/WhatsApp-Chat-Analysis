```python
import re 
import pandas as pd 
```

```python
f = open('WhatsApp Chat with Kusum Maity.txt','r', encoding='utf8')
```

```python
data = f.read()
```

```python
print(data)
```

```python
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
```

```python
messages = re.split(pattern,data)[1:] #taken from 1 as on 0th index an empty string was there
# len(messages)
messages
```

```python
dates = re.findall(pattern,data)
dates
```

```python
df = pd.DataFrame({'username':messages,'message_date':dates})
df.head()
```

```python
df['message_date'] = pd.to_datetime(df['message_date'], format = '%d/%m/%y, %H:%M - ')
```

```python
df.rename(columns={'message_date':'date'}, inplace=True)
```

```python
df.head()
```

```python
df.shape
```

```python
#separating user name and messages

users = []
messages = []
for message in df['username']:
    entry = re.split('([\\w\\W]+?):\\s', message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])
df['username(s)'] = users
df['message'] = messages
df.drop(columns=['username'], inplace=True)
```
